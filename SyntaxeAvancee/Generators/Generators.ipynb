```python
def finonacci():
    a, b = 0, 1
    while True:
        yield b
        a, b = b, a+b
        
fib = finonacci()
print(next(fib))
```

```python
def PrimeGenerator(limit):
    current = 2
    while current <= limit:
        is_prime = True
        for num in range(2, int(current**0.5)+1):
            if current % num == 0:
                is_prime = False
                break
                
        if is_prime:
            prime = current 
            current +=1
            yield prime
        current += 1
    #raise StopIteration        

for num in PrimeGenerator(20):
    print(num)
```

```python
def psychoilogue():
    print("Je vous ecoute")
    while True:
        answer = yield 
        if answer is not None:
            if answer.endswith("?"):
                print( "Question ?")
            else:
                print(" Phrase Affirmative")
bernard = psychoilogue()
next(bernard)
```

```python
def psychoilogue_execption():
    print("Je vous ecoute")
    while True:
        try : 
            answer = yield 
            if answer is not None:
                if answer.endswith("?"):
                    print( "Question ?")
                else:
                    print(" Phrase Affirmative")
        except Exception as e:
            print(f"Exception {e}")
        
bernard = psychoilogue_execption()
next(bernard)
bernard.send(" Vous allez bien ?")
bernard.send("Au revoir")
bernard.throw(Exception("Error Manifeste"))
```

```python
def psychoilogue_execption_close():
    print("Je vous ecoute")
    try :
        while True:
            try : 
                answer = yield 
                if answer is not None:
                    if answer.endswith("?"):
                        print( "Question ?")
                    else:
                        print(" Phrase Affirmative")
            except Exception as e:
                print(f"Exception {e}")
    except GeneratorExit:
        print("Au revoir le patient")
        
bernard = psychoilogue_execption_close()
#next(bernard)
bernard.send(" Vous allez bien ?")
bernard.send("Au revoir")
bernard.close()
```
